```python
#Ryan Kok
#Read values from a .csv file and then perform data calculations on them


import pandas as pd
import matplotlib.pyplot as chart

#Insert the values from the file into a pandas data frame
df = pd.read_csv('stockprice.csv')

print(df)

col = df["Amazon"]

#Find the max and min values for Amazon
print("\nMaximum stock price for Amazon is: " + str(col.max())) #(GeeksForGeeks, 2022)
print("\nMinimum stock price for Amazon is: " + str(col.min()))

col = df["Google"]

#Find the mean, variance and standard deviation for the Google values
print("\nThe mean of the Google stock prices is: " + str(col.mean()))
print("\nThe variance of the Google stock prices is: " + str(col.var()))
print("\nThe standard deviation of the Google stock prices is: " + str(col.std()))

#Create a line graph from the values
df.plot()

chart.show()

#Create a bar graph from the values
df.plot(kind="bar")

chart.show()
```
