```python
#Ryan Kok
#This program's function is to act as a GUI form to accept input and calculate the
#user's bill based on their inputs

import tkinter as Tk
from tkinter import *

#Class Billing START
#Each large group of code has comments around it to signal its start and end
class Billing():
    def __init__(self, master):
        self.master = master #Defining the attributes of the form's window
        master.title("Bill Calculator") #(Kumar, 2021)
        master.geometry("310x400")
        master.columnconfigure(1, weight=1)
        master.columnconfigure(2, weight=1)
        master.columnconfigure(3, weight=1)
        
        #Initiating attributes
        self.name = ""
        self.address = ""
        self.cellNo = 0
        self.baseCost = 0
        self.unitConsume = 0
        self.category = ""
        self.rate = 0
        self.bill = 0
        self.customerDetails = [""]
        
        
        #Intialising the text labels for each of the entries
        self.label = Label(master, text="Please input your details:")
        self.name_label = Label(master, text="Name:")
        self.address_label = Label(master, text="Address:")
        self.cell_label = Label(master, text="Cell No:")
        self.base_label = Label(master, text="Base Cost:")
        self.unit_label = Label(master, text="Units Consumed:")
        
        #Creating the entries themselves
        self.entryName = Entry(master)
        self.entryAddress = Entry(master)
        self.entryCell = Entry(master)
        self.entryBase = Entry(master)
        self.entryUnit = Entry(master)
        
        #Creating the radio button to select the category  (Klein, 2022)
        self.radioCate = StringVar()
        self.radioCategory1 = Radiobutton(master, text="Residential", value="Residential", variable = self.radioCate)
        self.radioCategory2 = Radiobutton(master, text="Commercial", value="Commercial", variable = self.radioCate)
        self.radioCategory3 = Radiobutton(master, text="Industrial", value="Industrial", variable = self.radioCate)
        
        #Creating the button that starts the validation process
        self.calc_button = Button(master, text="Calculate Bill", command=self.nameVali)
        
        #Creating the text widget for the validation checks and the output
        self.textbox = Text(master)
        
        #Places the instructions at the top of the form for the user
        self.label.grid(row=0, column=1, columnspan=3, sticky=N)
        
        #Placing the entry field labels on the grid
        self.name_label.grid(row=2, column=1)
        self.address_label.grid(row=3, column=1)
        self.cell_label.grid(row=4, column=1)
        self.base_label.grid(row=5, column=1)
        self.unit_label.grid(row=6, column=1)
        
        #Placing the entry fields on the grid
        self.entryName.grid(row=2, column=2, columnspan=2)
        self.entryAddress.grid(row=3, column=2, columnspan=2)
        self.entryCell.grid(row=4, column=2, columnspan=2)
        self.entryBase.grid(row=5, column=2, columnspan=2)
        self.entryUnit.grid(row=6, column=2, columnspan=2)
        
        #Place Radio Buttons on the grid
        self.radioCategory1.grid(row=7, column=1, columnspan=1, sticky=E)
        self.radioCategory2.grid(row=8, column=1, columnspan=1, sticky=E)
        self.radioCategory3.grid(row=9, column=1, columnspan=1, sticky=E)
        
        #Place calculate button
        self.calc_button.grid(row=10, column=1, columnspan=3, sticky=W+E, padx=5, pady=10)
        
        #Place textbox
        self.textbox.grid(row=11, column=1, columnspan= 3, pady=10)

    #Name Validation START
    #First of the validation for each of the fields. If the validation succeeds,
    #it calls the validation function of the next field. If it fails, the function returns
    #a null value, effectively stopping the program and allowing the user to retry their inputs
    
    def nameVali(self):
        self.textbox.delete("1.0", END) #(DelftStack, 2020)
        
        #Getting the user's name by getting the value from the appropriate entry box
        n = self.entryName.get()

        if len(n) > 0:
            #isaplha is a built-in python function that returns true if the variable only contains the alphabet
            if n.isalpha(): 
                self.name = n
                
                #Upon successful validation, a message is output to the form's textbox
                self.textbox.insert("1.0", "Name is valid.\n")
                
                #Calling the validation for the next entry field
                self.addressVali()
            else:
                #Upon unsuccessful validation, a message is output to the form's textbox
                self.textbox.insert("1.0", "Name has invalid character.\n")
                return
        else:
            self.textbox.insert("1.0", "Name cannot be left blank.\n")
            return
    #Name Validation END

    #Address Validation START
    def addressVali(self):
        a = self.entryAddress.get()

        if len(a) > 0:
            if a.isalpha():
                self.address = a
                self.textbox.insert("2.0", "Address is valid.\n")
                self.cellVali()
            else:
                self.textbox.insert("2.0", "Address has invalid characters.\n")
                return
        else:
            self.textbox.insert("2.0", "Address cannot be left blank\n")
            return
    #Address Validation END

    #Cell Number Validation START
    def cellVali(self):            
        c = self.entryCell.get()

        if len(c) == 10:
            #isnumeric is a built-in python function that returns true if the variable only contains numbers
            if c.isnumeric():
                self.cellNo = int(c)
                self.textbox.insert("3.0", "Cell Number is valid.\n")
                self.baseVali()
            else:
                self.textbox.insert("3.0", "Cell Number has invalid characters.\n")
                return
        else:
            self.textbox.insert("3.0", "Cell Number must be 10 digits long.\n")
            return
    #Cell Number Validation END

    #Base Cost Validation START
    def baseVali(self):
        b = self.entryBase.get()

        if len(b) > 0:
            if b.isnumeric():
                self.baseCost = int(b)
                self.textbox.insert("4.0", "Base Cost is valid.\n")
                self.unitVali()
            else:
                self.textbox.insert("4.0", "Base Cost has invalid characters.\n")
                return
        else:
            self.textbox.insert("4.0", "Base cost cannot be left blank.\n")
            return
    #Base Cost Validation END

    #Units Consumed Validation START
    def unitVali(self):
        u = self.entryUnit.get()

        if len(u) > 0:
            if u.isnumeric():
                self.unitConsume = int(u)
                self.textbox.insert("5.0", "Units Consumed is valid.\n")
                
                #Once all the validation checks have returned true, then the calculation can begin
                #First the rate is found based on the category chosen
                #Finally, the calculation function is called
                self.getRate()
                self.calculateBill()
            else:
                self.textbox.insert("5.0", "Units Consumed has invalid characters.\n")
                return
        else:
            self.textbox.insert("5.0", "Units consumed cannot be left blank.\n\n")
            return
    #Units Consumed Validation END

    #Get Rate per Category Function START
    def getRate(self):
        #The rates for each category is stored in a text file
        f = open("ratePerCategory.dat", "r")

        self.category = str(self.radioCate.get())
        
        #This loop checks if the category name is in the current line and then takes the rate
        # from the end of the line
        for i in f:
            place = i.find(self.category) #(GeeksForGeeks, 2021)
            if place > -1:
                place = i.find("=")
                self.rate = float(i[place + 1:len(i) - 1])

        f.close()
    #Get Rate per Category Function END

    #Calculate Bill Function START
    def calculateBill(self):
        #The bill calculation is baseCost + (UnitsConsumed * rate) + penalty
        #The rate is found in a text file and is based on which category was selected
        #The penalty is applied when the units consumed goes over the limit, which is different for each category
        #The penalty is an extra percentage that is applied to the units that are over the limit
            if self.category == "Residential":
                if self.unitConsume > 450:
                    extra = self.unitConsume - 450
                    self.bill = self.baseCost + (450 * self.rate) + (extra * self.rate) * 1.2
                else:
                    self.bill = self.baseCost + (self.unitConsume * self.rate)
            elif self.category == "Commercial":
                if self.unitConsume > 5000:
                    extra = self.unitConsume - 5000
                    self.bill = self.baseCost + (5000 * self.rate) + (extra * self.rate) * 1.3
                else:
                    self.bill = self.baseCost + (self.unitConsume * self.rate)
            elif self.category == "Industrial":
                if self.unitConsume > 30000:
                    extra = self.unitConsume - 30000
                    self.bill = self.baseCost + (30000 * self.rate) + (extra * self.rate) * 1.35
                else:
                    self.bill = self.baseCost + (self.unitConsume * self.rate)
                    
            #The details of the user are added to the list customerDetails
            self.setList        
            
            #The final bill is displayed in the text box
            total = "The Total Bill is : R" + str(self.bill)
            self.textbox.insert("6.0", total)
            
            
    #Calculate Bill Function END

    #Set List Function START
    #This function adds the customer details to the list customerDetails
    def setList(self):
        self.customerDetails[0] = self.name
        self.customerDetails[1] = self.address
        self.customerDetails[2] = self.cellNo
        self.customerDetails[3] = self.baseCost
        self.customerDetails[4] = self.unitConsume
        self.customerDetails[5] = self.bill            
    #Set List Function END

    #Class Billing END

#The main Billing class is assigned
#The tk mainloop is then called, telling the form to be displayed
gui = Tk()
main = Billing(gui)
gui.mainloop()
```
