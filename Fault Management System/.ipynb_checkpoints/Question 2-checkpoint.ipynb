```python
#Ryan Kok
#This program exists as a fault log management system, that allows an administrator to file a fault to a database
#and then view all the listed faults

import tkinter as Tk
from tkinter import *
from tkinter import ttk
from tkcalendar import DateEntry

import sqlite3
from sqlite3 import Error
import ctypes

class inputForm(): #Input Form Class Start
    def __init__(self, master):
        self.master = master #Defining the attributes of the form's window
        master.title("Fault Management")
        master.geometry("400x400")
        master.columnconfigure(1, weight=1)
        master.columnconfigure(2, weight=1)
        master.columnconfigure(3, weight=1)
        master.columnconfigure(4, weight=1)
        
        
        #Initiating attributes
        self.firstname = ""
        self.lastname = ""
        self.contact = ""
        self.apartment = ""
        self.unit = ""
        self.fault = ""
        
        #Intialising the text labels for each of the entries
        self.firstname_label = Label(master, text="First Name: ")
        self.lastname_label = Label(master, text="Last Name: ")
        self.contact_label = Label(master, text="Contact: ")
        self.apartment_label = Label(master, text="Apartment: ")
        self.reportdate_label = Label(master, text="Report Date: ")
        self.unit_label = Label(master, text="Unit: ")
        self.gender_label = Label(master, text="Gender: ")
        self.fault_label = Label(master, text="Fault: ")
        
        #Creating the entries themselves
        self.entryfirstname = Entry(master)
        self.entrylastname = Entry(master)
        self.entrycontact = Entry(master)
        self.entryapartment = Entry(master)
        self.entryunit = Entry(master)
        self.entryfault = Entry(master)
        
        #Creating and placing a date picker onto the grid (plus2net, 2021)
        self.reportdate = DateEntry(master, selectmode="day")
        self.reportdate.grid(row=5, column=2)
        
        #Placing the entry field labels on the grid
        self.firstname_label.grid(row=1, column=1)
        self.lastname_label.grid(row=2, column=1)
        self.contact_label.grid(row=3, column=1)
        self.apartment_label.grid(row=4, column=1)
        self.reportdate_label.grid(row=5, column=1)
        self.unit_label.grid(row=6, column=1)
        self.gender_label.grid(row=7, column=1)
        self.fault_label.grid(row=8, column=1)

        #Creating a radio group for the gender field
        self.radioGender = StringVar()
        self.radioGender.set("Male")
        self.radioGender1 = Radiobutton(master, text="Male", value="Male", variable = self.radioGender)
        self.radioGender2 = Radiobutton(master, text="Female", value="Female", variable = self.radioGender)
        
        #Placing the entry fields on the grid
        self.entryfirstname.grid(row=1, column=2, ipadx=50, pady=5, columnspan=2)
        self.entrylastname.grid(row=2, column=2, ipadx=50, pady=5, columnspan=2)
        self.entrycontact.grid(row=3, column=2, ipadx=50, pady=5, columnspan=2)
        self.entryapartment.grid(row=4, column=2, ipadx=50, pady=5, columnspan=2)
        self.entryunit.grid(row=6, column=2, ipadx=50, pady=5, columnspan=2)
        self.entryfault.grid(row=8, column=2, ipadx=50, ipady=20, pady=5, columnspan=2)
        
        #Creating the buttons for submission and listing faults
        self.submit_button = Button(master, text="Submit Now", command=self.firstnameVali)
        self.listFault_button = Button(master, text="List Fault", command=self.displayList)
        
        #Placing the radio buttons on the grid
        self.radioGender1.grid(row=7, column=2)
        self.radioGender2.grid(row=7, column=3)
        
        #Placing the buttons on the grid
        self.submit_button.grid(row=9, column= 3)
        self.listFault_button.grid(row=9, column=2)
        
        
    def firstnameVali(self):   
        #Getting the user's first name by getting the value from the appropriate entry box
        n = self.entryfirstname.get()

        if len(n) > 0:
            #isaplha is a built-in python function that returns true if the variable only contains the alphabet
            if n.isalpha(): 
                self.firstname = n

                #Calling the validation for the next entry field
                self.lastnameVali()
            else:
                #Upon unsuccessful validation, a message is output to the screen

                #Code that creates a windows pop up message to dispaly an error
                ctypes.windll.user32.MessageBoxW(0, "First name has invalid character.", "Error", 0) 
                return
        else:
            ctypes.windll.user32.MessageBoxW(0, "First name cannot be left blank.", "Error", 0) 
            return
        
    def lastnameVali(self):   
        #Getting the user's last name by getting the value from the appropriate entry box
        n = self.entrylastname.get()

        if len(n) > 0:
            if n.isalpha(): 
                self.lastname = n

                #Calling the validation for the next entry field
                self.contactVali()
            else:
                #Upon unsuccessful validation, a message is output to the screen

                ctypes.windll.user32.MessageBoxW(0, "Last name has invalid character.", "Error", 0) 
                return
        else:
            ctypes.windll.user32.MessageBoxW(0, "Last name cannot be left blank.", "Error", 0) 
            return
        
    def contactVali(self):            
        c = self.entrycontact.get()

        if len(c) == 10:            
            if c.isnumeric(): 
                self.contact = int(c)
                
                self.apartmentVali()
            else:
                ctypes.windll.user32.MessageBoxW(0, "Cell Number has invalid characters.", "Error", 0) 
                return
        else:
            ctypes.windll.user32.MessageBoxW(0, "Cell Number must be 10 digits long", "Error", 0) 
            return
    
    def apartmentVali(self):   
        #Getting the user's apartment by getting the value from the appropriate entry box
        n = self.entryapartment.get()

        if len(n) > 0:
            if n.isalpha(): 
                self.apartment = n

                #Calling the validation for the next entry field
                self.unitVali()
            else:
                #Upon unsuccessful validation, a message is output to the screen

                ctypes.windll.user32.MessageBoxW(0, "Apartment field has invalid character.", "Error", 0) 
                return
        else:
            ctypes.windll.user32.MessageBoxW(0, "Apartment field cannot be left blank.", "Error", 0) 
            return

        
    def unitVali(self):            
        c = self.entryunit.get()

        if len(c) > 0:
            #isnumeric is a built-in python function that returns true if the variable only contains numbers
            if len(c) > 4:
                ctypes.windll.user32.MessageBoxW(0, "Unit Number cannot be longer than 4 digits.", "Error", 0) 
                return
            elif c.isnumeric():
                self.unit = int(c)
                self.fault = self.entryfault.get()
                
                #Calls the dataOperation class to input the data into the database after validation
                db = dataOperation()
                db.createTable()
                db.inputData(self.firstname, self.lastname, self.contact, self.apartment, self.reportdate.get(), self.unit, self.radioGender.get(), self.fault)
                del db
                
                ctypes.windll.user32.MessageBoxW(0, "Fault has been submitted.", "Submission", 0) 
            else:
                ctypes.windll.user32.MessageBoxW(0, "Unit Number has invalid characters.", "Error", 0) 
                return
        else:
            ctypes.windll.user32.MessageBoxW(0, "Unit Number cannot be blank.", "Error", 0) 
            return
        
    def displayList(self):
        listFaults = listForm(self.master)
        listFaults.displayData()
        
   #Input Form Class End     
        
class listForm(): #List Form Class Start
    def __init__(self, master):
        newWindow = Toplevel(master) #Creating a new window using Toplevel (geeksforgeeks, 2021)
        self.master = master
        newWindow.title("Faults List")
        newWindow.geometry("900x200")
        
        #Creating a treeview to display all the faults on a new window (Kumar, 2021)
        self.table = ttk.Treeview(newWindow)
        
        self.table["columns"] = ("ID", "FirstName", "LastName", "Contact", "Apartment", "ReportDate", "Unit", "Gender", "Fault")
        
        #Defining all the columns
        self.table.column("#0", width=0, stretch=NO)
        self.table.column("ID", anchor=CENTER, width=30)
        self.table.column("FirstName", anchor=W, width=100)
        self.table.column("LastName", anchor=W, width=100)
        self.table.column("Contact", anchor=W, width=100)
        self.table.column("Apartment", anchor=W, width=150)
        self.table.column("ReportDate", anchor=CENTER, width=100)
        self.table.column("Unit", anchor=W, width=30)
        self.table.column("Gender", anchor=CENTER, width=80)
        self.table.column("Fault", anchor=W, width=200)
        
        #Naming all the headings
        self.table.heading("#0", text="", anchor=CENTER)
        self.table.heading("ID", text="ID", anchor=W)
        self.table.heading("FirstName", text="First Name", anchor=W)
        self.table.heading("LastName", text="Last Name", anchor=W)
        self.table.heading("Contact", text="Contact", anchor=W)
        self.table.heading("Apartment", text="Apartment", anchor=CENTER)
        self.table.heading("ReportDate", text="Report Date", anchor=CENTER)
        self.table.heading("Unit", text="Unit", anchor=W)
        self.table.heading("Gender", text="Gender", anchor=CENTER)
        self.table.heading("Fault", text="Fault", anchor=W)
        
        self.table.pack()
        
    def displayData(self):
        #Create a database file if one does not already exist
        file_path = r"FaultDatabase.db"
        
        try:
            #Creating a connection to the database file
            self.conn = sqlite3.connect(file_path)
        except Error:
            ctypes.windll.user32.MessageBoxW(0, "Connection Creation Error", "Error", 0)
            
        self.curs = self.conn.cursor()
        
        sql = """SELECT * FROM faults"""
        
        self.curs.execute(sql)
        
        data = self.curs.fetchall()
        
        #Displaying the data onto the treeview
        for i in data:
            self.table.insert("", END, values=i)
       
        self.conn.close()
    #List Form Class End
    
class dataOperation():    #dataOperation Class Start
    def __init__(self):
        file_path = r"FaultDatabase.db"
        
        try:
            self.conn = sqlite3.connect(file_path)
        except Error:
            ctypes.windll.user32.MessageBoxW(0, "Connection Creation Error", "Error", 0)
            
        self.curs = self.conn.cursor()
        
    
    def createTable(self):
        
        #SQL script that will create a table if one does not already exist
        sql = """CREATE TABLE IF NOT EXISTS faults
            (ID integer PRIMARY KEY,
            firstname text NOT NULL,
            lastname text NOT NULL,
            contact text NOT NULL,
            apartment text NOT NULL,
            reportdate text NOT NULL,
            unit integer NOT NULL,
            gender text NOT NULL,
            fault text NOT NULL
            )
            """
        try:
            self.curs.execute(sql)
        except:
            ctypes.windll.user32.MessageBoxW(0, "Table Creation Error", "Error", 0)
    
    def inputData(self, firstname, lastname, contact, apartment, reportdate, unit, gender, fault):
        
        #Selects the last row in the table, so that we can use the most recent ID to make the next ID
        # (TutorialsPoint, 2018)
        self.curs.execute("SELECT ID FROM faults ORDER BY ID DESC LIMIT 1")
        
        try:
            rawID = self.curs.fetchone()
            ID = rawID[0]
            ID = ID + 1
        except:
            ID = 1

        sql = """INSERT INTO faults(ID, firstname, lastname, contact, apartment, reportdate, unit, gender, fault)
              VALUES(?, ?, ?, ?, ?, ?, ?, ?, ?)"""
        values = (ID, firstname, lastname, contact, apartment, reportdate, unit, gender, fault)
        
        self.curs.execute(sql, values)
        self.conn.commit()
        self.conn.close()
    #dataOperation Class End
    
gui = Tk()
main = inputForm(gui)
gui.mainloop()

```
